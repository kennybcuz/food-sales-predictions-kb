# Visualizations

## Histograms

```python
# histgram of sales data by Tier

# set plt figure size
plt.figure(figsize=(15,10))

```

```python
# Boxplot of sales by Item Type


# set plt figure size
plt.figure(figsize=(20,10))
plt.xticks(rotation=90, fontsize=18)
plt.yticks(fontsize=18)

sb.boxplot(data=df, x='Item_Type',y='Item_Outlet_Sales', linewidth= 2);
```

```python
# histgram of sales by Fat Content
# low fat has more sales

# set plt figure size
plt.figure(figsize=(15,10))

sb.histplot(data=df, x='Item_Outlet_Sales', hue='Item_Fat_Content', element='step');
```

```python
# Boxplot of item type by Item visibility


# set plt figure size
plt.figure(figsize=(20,10))
plt.xticks(rotation=90, fontsize=18)
plt.yticks(fontsize=18)

sb.boxplot(data=df, x='Item_Type',y='Item_Visibility', linewidth= 2);
```

```python
# Boxplot of Item Sales by Outlet Identifier


# set plt figure size
plt.figure(figsize=(20,10))
plt.xticks(rotation=90, fontsize=18)
plt.yticks(fontsize=18)

sb.boxplot(data=df, x='Outlet_Identifier',y='Item_Outlet_Sales', linewidth= 2, hue='Item_Fat_Content');
```
